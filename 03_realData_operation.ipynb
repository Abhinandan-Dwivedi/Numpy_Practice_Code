```python
import numpy as np
```

```python
                # id, 2021, 2022, 2023, 2024,2025
data = np.array([[1,10000,20000,30000,40000,50000],   # vaisno dhaba
                [2,90000,60000,5000,34000,21000],     # kfc
                [3,4400,5500,6600,77000,78000],       # swiggy
                [4,12000,130000,14000,15000,16000]])  # zomato
```

### sum of the price per year

```python
sum = np.sum(data , axis = 0)
print("sum with id : ",sum)

sum = np.sum(data[: ,1:] , axis = 0 )
print("sum without id :" , sum)
```

### min and max price of restorant

```python
# minimum price of every retorent 
print (" min price of Every  restorent" , np.min(data[:, 1:] , axis = 1))

# Maximun 
print (" min price of Every  restorent" , np.max(data[:, 1:] , axis = 1))

# maximun per year
print (" min price of Every  restorent" , np.max(data , axis = 0))
```

### mean an cumulatiove sum

```python
# Average sale of every restaurent 
print(np.mean(data[:,1:] , axis = 1))

# cumulative sum of every restaurent
print(np.cumsum(data[:,1:] , axis = 1))

import matplotlib.pyplot as plt
plt.plot(np.mean(data[:,1:] , axis = 1))
plt.xlabel("restaurent")
plt.ylabel("price")
plt.title("Average price of restaurent")
plt.show()
```

### Operations as a vector 

```python
vec1 = np.array([2,3,4,5,6]) 
vec2 = np.array([1,2,3,4,5])
print(" multiplication : " , vec1 * vec2)

# scaler multiplication (dot product)
print(np.dot(vec1, vec2))

# angle bw two vector 
angle = np.arccos(np.dot(vec1, vec2)/(np.linalg.norm(vec1) * np.linalg.norm(vec2)))
print(angle)
```
